# MATPLOTLIB



```python
import matplotlib.pyplot as plt
```

# Line Graph

```python
x= [1,2,4,5,6,8]
y= [2,5,7,9,11,13]
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,marker='o')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.title("Graph")
plt.show()
```

# BAR GRAPH

```python
x= [1,2,4,5,6,8]
y= [2,5,7,9,11,13]

c=['r','g','b','y','c','k']
plt.barh(x,y,color= c)
plt.show()
```

# Scatter Plot

```python
x= [1,2,4,5,6,8]
y= [2,5,7,9,11,13]

c=['r','g','b','y','c','k']
plt.scatter(x,y,color= c,s=150)
plt.show()
```
